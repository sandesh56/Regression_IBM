```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
''' This is the interpretation of simple linear example
having equation 2X^2+ 3 '''
x = np.arange(-5.0, 5.0, 0.1) #nitialize an array from -5 to 5 with increament of 0.1

##You can adjust the slope and intercept to verify the changes in the graph
y = 2*(x) + 3
noise= 2 * np.random.normal(size=x.size) ###Draw random samples from a normal (Gaussian) distribution.
data = y + noise
#plt.figure(figsize=(8,6))
plt.plot(x, data,  'go') ### 'bo' display with blue and "go" display with green circle.
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

```python
''' Interpretation of complex regression'''

x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph
y = 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)###Draw random samples from a normal (Gaussian) distribution.
ydata = y + y_noise
plt.plot(x, ydata,  'bo') ### 'bo' display with blue and "go" display with green circle.
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

```python
#Quadratic function i.e Y = X^2

x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

y = np.power(x,2)
noise = 2 * np.random.normal(size=x.size)
data = y + noise
plt.plot(x, data,  'go')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

```python
#exponential function 

X = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

Y= np.exp(X)
noise = 2* np.random.normal(size=X.size)
data = Y + noise
plt.plot(X,data, 'bo')

plt.plot(X,Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```
